```python
from docking.dock import MolDox
from pymol import cmd
from rdkit import Chem
from docking.utils import split_pdb
```

```python
output_dir = 'test_output'
receptor = f'{output_dir}/1AZ8_clean.pdb'
ref_ligand = f'{output_dir}/1AZ8_lig.mol2'

# In this example, we are re-docking a ligand, but this file could be any mol2 or sdf file
docking_ligands = f'{output_dir}/1AZ8_lig.mol2'
```

```python
split_pdb(receptor, ref_ligand, fetch='1AZ8')
```

```python
dox = MolDox(output_dir)
dox.prep_receptor(receptor, ref_ligand, ph=7.4, keep_water=True, renumber=True)
```

```python
dox.prep_ligands(docking_ligands, format='mol2')
```

```python
dox.autodock_all()
```

```python
results=Chem.SDMolSupplier(dox.docking_results['mol_0'])
p=Chem.MolToMolBlock(results[0],False)
```

```python
dox.viewer()
dox.view.add_layer_from_object(p)
dox.view.sticks(colorscheme='cyanCarbon')
dox.show()
```
