```python
from docking.dock import MolDox
from pymol import cmd
from rdkit import Chem
```

```python
cmd.fetch(code='1AZ8',type='pdb1')
cmd.select(name='Prot',selection='polymer.protein')
cmd.select(name='Lig',selection='organic')
cmd.save(filename='test_output/1AZ8_clean.pdb',format='pdb',selection='Prot')
cmd.save(filename='test_output/1AZ8_lig.mol2',format='mol2',selection='Lig')
cmd.delete('all')
```

```python
dox = MolDox(output_dir='test_output')
dox.prep_receptor(receptor='test_output/1AZ8_clean.pdb',
                  ligand='test_output/1AZ8_lig.mol2',
                  ph=7.4, keep_water=True, renumber=True)
```

```python
dox.prep_ligands('test_output/1AZ8_lig.mol2', format='mol2')
```

```python
dox.autodock_all()
```

```python
results=Chem.SDMolSupplier(dox.docking_results['mol_0'])
p=Chem.MolToMolBlock(results[0],False)
```

```python
dox.viewer()
dox.view.add_layer_from_object(p)
dox.view.sticks(colorscheme='cyanCarbon')
dox.show()
```
